# Book Recommender System


```python
import numpy as np
import pandas as pd
```

```python
books= pd.read_csv('Books.csv')
users = pd.read_csv('Users.csv')
ratings = pd.read_csv('Ratings.csv')

```

```python
books.head()
```

```python
ratings.head()
```

```python
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_rating']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','num_rating','avg_rating' ]]
```

```python
popular_df
```

# # Collaborative filtering system


```python
x= ratings_with_name.groupby('User-ID').count()['Book-Rating'] >200
read_user= x[x].index
```

```python
 filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(read_user)]
```

```python
y = ratings_with_name.groupby('Book-Title').count()['Book-Rating'] >=50
famous_books= y[y].index
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_rating.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
similarity_scores = cosine_similarity(pt)
```

# #Function of Search in book 

```python
def recomend(book_name):
    index= np.where(pt.index == book_name)[0][0]
    similar_item =sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1] ,reverse=True) [1:6]
    for i in similar_item:
        print(pt.index[i[0]])
```
